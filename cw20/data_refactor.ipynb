# This file is irrelevant for data analysis purposes!
It merely refactors the data to use a naming convention and tell which data is garbage and which is usable.

```python
filenames = [
"fotoczerwony",
"fotoniebieska",
"fotonieos",
# "fotonieosw",
# "fotonieoswi",
# "fotonieoswietlona",
# "krzem",
# "krzemdobry",
"krzemdobry2",
# "krzemsus",
# "schottky",
"Szotkydobry",
"Tunelowa",
"zener",
]

dsets = {}
```

```python
for filename in filenames:
    filepath = "in/"+filename
    dsets[filename] = pd.read_csv(filepath, sep='\s+')
    dsets[filename] = dsets[filename].rename(columns = {"U[V]":"U", "I[A]":"I"})

dsets['fotoczerwony']
```

```python
for key in dsets:
    dsets[key].plot.scatter(x='U', y='I', title=key)
```

```python
dsets['photored'] = dsets.pop("fotoczerwony")
dsets['photoblue'] = dsets.pop("fotoniebieska")
dsets['photodark'] = dsets.pop("fotonieos")
dsets['silicon'] = dsets.pop("krzemdobry2")
dsets['schottky'] = dsets.pop("Szotkydobry")
dsets['tunnel'] = dsets.pop("Tunelowa")
dsets['zener'] = dsets.pop("zener")
```

```python
for key in dsets:
    filepath = "in_refactored/" + key + ".txt"
    dsets[key].to_csv(filepath, index = False, sep = '\t')
```
