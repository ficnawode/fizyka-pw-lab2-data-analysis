```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cds = pd.read_csv("in/CdS_500um", sep='\s+')
gap = pd.read_csv("in/GaP_400um", sep='\s+')
I0 = pd.read_csv("in/I_0", sep='\s+')
```

```python
plt.figure(figsize=(8,6))

dotcolor = "orange"
plt.plot(cds['lambda'], cds['I'], color = dotcolor)

plt.xlabel(r"$\lambda \; [nm]$")
plt.ylabel(r"$I \; [a.u.]$")

plt.savefig("out/Cds_I_od_lambda.jpg", dpi=350)
plt.show()
```

```python

plt.figure(figsize=(8,6))

dotcolor = "magenta"
plt.plot(gap['lambda'], gap['I'], color = dotcolor)

plt.xlabel(r"$\lambda \; [nm]$")
plt.ylabel(r"$I \; [a.u.]$")

plt.savefig("out/GaP_I_od_lambda.jpg", dpi=350)
plt.show()
```

```python
plt.figure(figsize=(8,6))

dotcolor = "purple"
plt.plot(I0['lambda'], I0['I'], color = dotcolor)

plt.xlabel(r"$\lambda \; [nm]$")
plt.ylabel(r"$I \; [a.u.]$")

plt.savefig("out/I0_I_od_lambda.jpg", dpi=350)
plt.show()
```
