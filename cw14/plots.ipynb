```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#global constants
_plot_width = 11
_plot_height = 8
_dotsize = 30
_linewidth = 0.5
_dpi = 350
```

```python

T = pd.read_csv("csv/T.csv")
P = pd.read_csv("csv/P.csv")

T_short = T[0:-4]

plt.figure(figsize=(_plot_width,_plot_height))

dotcolor = "orange"
plt.scatter(T_short["T"], P["P"], s = _dotsize, color=dotcolor)
plt.errorbar(T_short["T"], P["P"], xerr=T_short["u"],yerr=P["u"],capsize=3,capthick=_linewidth,elinewidth=_linewidth,fmt="none", ecolor="black")

plt.xlabel(r"$T$ $[K]$")
plt.ylabel(r"$P$ $\left[\dfrac{C}{m^2}\right]$")

plt.savefig("img/wyk_P_od_T.jpg", dpi=_dpi)
plt.show()
```

```python

Ec = pd.read_csv("csv/Ec.csv")

plt.figure(figsize=(_plot_width,_plot_height))

dotcolor = "green"
plt.scatter(T_short["T"], Ec["Ec"], s = _dotsize, color=dotcolor)
plt.errorbar(T_short["T"], Ec["Ec"], xerr=T_short["u"],yerr=Ec["u"],capsize=3,capthick=_linewidth,elinewidth=_linewidth,fmt="none", ecolor="black")

plt.xlabel(r"$T$ $[K]$")
plt.ylabel(r"$E_c$ $\left[\dfrac{V}{m}\right]$")

plt.savefig("img/wyk_Ec_od_T.jpg", dpi=_dpi)
plt.show()
```

```python
eps = pd.read_csv("csv/eps.csv")

plt.figure(figsize=(_plot_width,_plot_height))

dotcolor = "blue"
plt.scatter(T["T"], eps["eps"], s = _dotsize, color=dotcolor)
plt.errorbar(T["T"], eps["eps"], xerr=T["u"],yerr=eps["u"],capsize=3,capthick=_linewidth,elinewidth=_linewidth,fmt="none", ecolor="black")

plt.xlabel(r"$T$ $[K]$")
plt.ylabel(r"$\varepsilon$")

plt.savefig("img/wyk_eps_od_T.jpg", dpi=_dpi)
plt.show()
```

```python
one_over_chi = pd.read_csv("csv/one_over_chi.csv")

plt.figure(figsize=(_plot_width,_plot_height))

dotcolor = "pink"
plt.scatter(T["T"], one_over_chi["one_over_chi"], s = _dotsize, color=dotcolor)
plt.errorbar(T["T"], one_over_chi["one_over_chi"], xerr=T["u"],yerr=one_over_chi["u"],capsize=3,capthick=_linewidth,elinewidth=_linewidth,fmt="none", ecolor="black")

plt.xlabel(r"$T$ $[K]$")
plt.ylabel(r"$\dfrac{1}{\chi}$")

plt.savefig("img/wyk_one_over_chi_od_T.jpg", dpi=_dpi)
plt.show()
```
