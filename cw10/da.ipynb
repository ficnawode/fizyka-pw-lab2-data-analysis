```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
char = pd.read_csv('in/pomiary5.dat', sep='\s+')
char['U'] /= 1000 #convert mV to V

ref1 = pd.read_csv('in/reference/CH18mA.txt', sep='\s+', decimal = ',',names=['t', 'U'])
ref1['t'] /= 20

ref2 = pd.read_csv('in/reference/DCH18mA.txt', sep='\s+', decimal = ',',names=['t', 'U'])
ref2['t'] /= 20
ref2['t'] += ref1['t'].max()
```

```python
plt.figure(figsize=(8,6))

plt.plot(char['t'], char['U'], color = 'orange', label = 'Eksperyment',linewidth = 1)
plt.plot(ref1['t'], ref1['U'], color = 'red',label = 'Teoria', linewidth = 1)
plt.plot(ref2['t'], ref2['U'], color = 'red', linewidth = 1)

plt.xlabel(r'$t \; [s]$')
plt.ylabel(r'$U \; [V]$')

plt.legend()

plt.savefig('out/wyk_U_od_t.jpg', dpi = 350)
plt.show()
```
