```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
char = pd.read_csv('in/pomiary5.dat', sep='\s+')
char['U'] /= 1000 #convert mV to V

ref1 = pd.read_csv('in/reference/CH18mA.txt', sep='\s+', decimal = ',',names=['t', 'U'])
ref1['t'] /= 20

ref2 = pd.read_csv('in/reference/DCH18mA.txt', sep='\s+', decimal = ',',names=['t', 'U'])
ref2['t'] /= 20
ref2['t'] += ref1['t'].max()
```

```python
plt.figure(figsize=(8,6))

plt.plot(char['t'], char['U'], color = 'orange', label = 'Eksperyment',linewidth = 1)
plt.plot(ref1['t'], ref1['U'], color = 'red',label = 'Teoria', linewidth = 1)
plt.plot(ref2['t'], ref2['U'], color = 'red', linewidth = 1)

plt.xlabel(r'$t \; [s]$')
plt.ylabel(r'$U \; [V]$')

plt.legend()

plt.savefig('out/wyk_U_od_t.jpg', dpi = 350)
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,6))
midway_point = 1545
char1 = char[:midway_point]
ax[0].plot(char1['t'], char1['U'], color = 'green', label = 'Eksperyment',linewidth = 1)
ax[0].set_title('krzywa naadowania')
ax[0].set_xlabel(r'$t \; [s]$')
ax[0].set_ylabel(r'$U \; [V]$')

char1 = char[midway_point:].copy()
char1['t'] -= char1['t'].min()
ax[1].plot(char1['t'], char1['U'], color = 'purple', label = 'Eksperyment',linewidth = 1)
ax[1].set_title('krzywa rozadowania')
ax[1].set_xlabel(r'$t \; [s]$')
ax[1].set_ylabel(r'$U \; [V]$')

plt.savefig('out/charakterystyki.jpg', dpi = 350)
plt.show()
```
